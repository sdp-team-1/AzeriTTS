```
from pathlib import Path
import shutil
import os

for path in Path(os.getcwd()).rglob('*.py'):
  flis = list()
  with open(path,"r") as f:
    flis = f.readlines()
 
  rep = open(path,"w")
  for fl in flis:
    nuli = fl
    if "examples." in fl:
      nuli = nuli.replace("examples.","ttsexamples.")
      print(nuli.strip())
 
 
    rep.write(nuli)
 
  rep.close()
 
shutil.move("examples","ttsexamples")
```

```
!pip install PyDrive
!pip install pydub

from pydrive.auth import GoogleAuth
from pydrive.drive import GoogleDrive
from os import listdir
from os.path import isfile, join
from pydub import AudioSegment

audio_files = [f for f in listdir('../SDP/Data/wavs/') if isfile(join('../SDP/Data/wavs/', f))]

for idx, file_path in enumerate(audio_files):
    print('Progress: {}'.format(idx / len(audio_files)))
    sound = AudioSegment.from_wav('../SDP/Data/wavs/' + file_path)
    sound = sound.set_channels(1)
    sound.export('../SDP/Data/wavs/' + file_path, format="wav")
```
